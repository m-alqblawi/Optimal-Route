## 04.01 DBSCAN

```python
from sklearn.cluster import DBSCAN

dbscan_model = DBSCAN(eps=0.07, min_samples=2, algorithm='auto', metric='cityblock').fit(df[['lng','lat']])
labels = dbscan_model.labels_
```

```python
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)
print(f"number of cluster: {n_clusters}")
print(f"number of noise in label: {n_noise}")
```

```python
lng = df['lng']
lat = df['lat']
labels = dbscan_model.labels_

# Create a scatter plot for the clusters
plt.scatter(lng[labels != -1], lat[labels != -1], c=labels[labels != -1], cmap='viridis', label='Clusters')
plt.colorbar()

# Create a scatter plot for the noise points
plt.scatter(lng[labels == -1], lat[labels == -1], c='red', marker='x', label='Noise')

# Set plot title and labels
plt.title('DBSCAN Clustering')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

## 04.02 OPTICS

```python
from sklearn.cluster import OPTICS

optics_model = OPTICS(min_samples=2, algorithm='auto', metric='cityblock')
labels_optics = optics_model.fit_predict(df[['lng', 'lat']])

# Get the number of clusters (excluding noise)
n_clusters_optics = len(set(labels_optics)) - (1 if -1 in labels_optics else 0)
n_noise_optics = list(labels_optics).count(-1)

print(f"Number of clusters: {n_clusters_optics}")
print(f"Number of noise points: {n_noise_optics}")

# Create a scatter plot for the clusters
plt.scatter(df['lng'][labels_optics != -1], df['lat'][labels_optics != -1], c=labels_optics[labels_optics != -1], cmap='viridis', label='Clusters')
plt.colorbar()

# Create a scatter plot for the noise points
plt.scatter(df['lng'][labels_optics == -1], df['lat'][labels_optics == -1], c='red', marker='x', label='Noise')

# Set plot title and labels
plt.title('OPTICS Clustering')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

## 04.03 KMean

```python
from sklearn.cluster import KMeans
from scipy.spatial import Voronoi, voronoi_plot_2d

kmeans = KMeans(n_clusters=27)
kmeans.fit(df[['lng','lat',  ]])
cluster = kmeans.predict(df[['lng','lat', ]])
sns.scatterplot(x=df['lng'],y=df['lat'],hue=cluster,palette='rainbow',legend=False)

centers=kmeans.cluster_centers_
vor = Voronoi(centers)
fig = voronoi_plot_2d(vor,plt.gca())

plt.show()
```

```python
df_train = df.copy()
df_train
```

```python
from sklearn.cluster import KMeans
import pandas as pd
YOUR_GOVERNATE_CLUSTER_COUNT = 27
YOUR_AREA_CLUSTER_COUNT = 3
# Assuming you have a DataFrame df_train with columns "governate", "area", "lat", "lng", and "services"

# Level 1: Cluster by Governorate
unique_governates = df_train["governate"].unique()

governate_clusters = {}  # Dictionary to store clusters by governorate

for governate in unique_governates:
    governate_df = df_train[df_train["governate"] == governate]
    
    # Perform K-Means clustering for governorate based on lat and lng
    kmeans_governate = KMeans(n_clusters=YOUR_GOVERNATE_CLUSTER_COUNT)
    governate_df["governate_cluster"] = kmeans_governate.fit_predict(governate_df[["lng", "lat"]])
    
    governate_clusters[governate] = governate_df

# Level 2: Cluster by Area within Each Governorate
area_clusters = {}  # Dictionary to store clusters by area within each governorate

for governate, governate_df in governate_clusters.items():
    unique_areas = governate_df["area"].unique()
    
    area_clusters[governate] = {}  # Dictionary to store area clusters within each governorate
    
    for area in unique_areas:
        area_df = governate_df[governate_df["area"] == area]
        
        # Perform K-Means clustering for area based on lat and lng
        kmeans_area = KMeans(n_clusters=YOUR_AREA_CLUSTER_COUNT)
        area_df["area_cluster"] = kmeans_area.fit_predict(area_df[["lng", "lat"]])
        
        area_clusters[governate][area] = area_df

# Now, you have clusters by governorate and area within each governorate.
# You can access the clusters using governate_clusters and area_clusters dictionaries.
```
